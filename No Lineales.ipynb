### Sanitizar entradas, validar, checar si podemos hacer la gráfica interactiva

```python
x,y,z = symbols("x y z")
funcion = input('Ingrese la función: ')
sympify(funcion)
```

```python
simbolica = sympify(funcion)
grado = degree(simbolica) #obtener grado de la función si el rango especificado por el usuario no basta para encontrar todas las raíces
menor = float(input("Ingrese el límite inferior del rango a evaluar: "))
mayor = float(input("Ingrese el límite mayor del rango a evaluar: "))
rango = np.arange(menor, (mayor+1))
f = lambdify(x, simbolica, "numpy")  # crea una expresion de numpy para evaluar
resultado = f(rango)  # evalua en el rango seleccionado
signs = np.sign(resultado)  # para obtener donde cruza 0
signs[signs == 0] = -1
zero_crossings = np.where(np.diff(signs))[0]
cambio = []  # valores del cambio de signo
anterior_cambio = []  # valor anterior al cambio de signo


for a in range(len(zero_crossings)):
    temporal = zero_crossings[a]
    anterior_cambio.append(resultado[temporal])
    cambio.append(resultado[temporal+1])


rango_nuevo = []
while(len(zero_crossings) < grado):
    print("El rango especificado no ha encontrado todas las raíces de la función, aumentando el rango ")
    menor = menor * 5
    mayor = mayor * 5
    rango = np.arange(menor, mayor)
    f = lambdify(x, simbolica, "numpy")  # crea una expresion de numpy para evaluar
    resultado = f(rango)  # evalua en el rango seleccionado
    signs = np.sign(resultado)  # para obtener donde cruza 0
    signs[signs == 0] = -1
    zero_crossings = np.where(np.diff(signs))[0]
    cambio = []  # valores del cambio de signo
    anterior_cambio = []  # valor anterior al cambio de signo
    for a in range(len(zero_crossings)):
        temporal = zero_crossings[a]
        anterior_cambio.append(resultado[temporal])
        cambio.append(resultado[temporal+1])

        
print("Rango establecido de",menor,"a",mayor)
for a in range(0, len(zero_crossings)):
    print("Para x en: ", rango[zero_crossings[a]],
          ", f(x) vale: ", anterior_cambio[a])
    print("Para x en: ", rango[zero_crossings[a]+1],
          ", f(x) vale: ", cambio[a], " cambio de signo")
    rango_nuevo.append(rango[zero_crossings[a]+1])

positivos = []  # para bisectriz
negativos = []  # para bisectriz

for i in range(int(menor), int(mayor)):
    if f(i) > 0:
        if f(i) in cambio:
            positivos.append(i)
        elif f(i) in anterior_cambio:
            positivos.append(i)
    else:
        if f(i) in cambio:
            negativos.append(i)
        elif f(i) in anterior_cambio:
            negativos.append(i)


fig = plt.figure()
ax = plt.gca()  # get current axis
ax.plot(rango, resultado)
ax.grid(True)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
plt.plot(rango_nuevo, cambio, 'ro')
plt.show()
```

## Bisección

```python
for i in range(len(negativos)):
    a = positivos[i] #a siempre es donde se comporta positivo
    b = negativos[i] #b siempre es donde se comporta negativo
    root = optimize.bisect(f,a,b)
    print("La raíz entre a:", a,"y b:", b, "es:", root)
```
